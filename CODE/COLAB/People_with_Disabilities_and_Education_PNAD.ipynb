## DATA VIS I
### First, I wanted to understand the general level. Therefore, I used the Total data that sums up all the education levels of people with disabilities for each region of the country. I used the Altair library to create a scatter plot.

```python
import altair as alt

# Enable Altair rendering mode for Google Colab
alt.renderers.enable('colab')

# Filter relevant columns for education levels
niveis_instrucao = ['sem instrução', 'nível básico', 'nível médio', 'nível superior']
colunas_selecionadas = [f'Pessoa {nivel_instrucao} com deficiência' for nivel_instrucao in niveis_instrucao]

# Filter data for relevant regions and conditions
filtro_linhas = tabela3['Localidade'].isin(['Norte', 'Nordeste', 'Centro-Oeste', 'Sul', 'Sudeste']) & \
                tabela3['Condição em relação à força de trabalho'].isin(['Na força de trabalho', 'Fora da força de trabalho'])
dados_plot = tabela3[['Localidade', 'Condição em relação à força de trabalho'] + colunas_selecionadas][filtro_linhas]

# Create a column for the total number of people per region
dados_plot['Total'] = dados_plot.groupby('Localidade')[colunas_selecionadas].transform('sum').sum(axis=1)


# Transform the data into the appropriate format for Altair
data = {
    'Localidade': dados_plot['Localidade'],
    'Força de Trabalho': dados_plot['Condição em relação à força de trabalho'],
    'Pessoa com deficiência': dados_plot[colunas_selecionadas].sum(axis=1),
    'Total': dados_plot['Total']
}
df = pd.DataFrame(data)

# Filter the data for employment conditions
filtro_laranja = df['Força de Trabalho'] == 'Na força de trabalho'
filtro_azul = df['Força de Trabalho'] == 'Fora da força de trabalho'

# Map formats for employment conditions
formato_condicoes = alt.Shape('Força de Trabalho:N', legend=alt.Legend(title='Legenda'),
                              scale=alt.Scale(domain=['Na força de trabalho', 'Fora da força de trabalho'],
                                              range=['triangle-up', 'square']))

# Create an interactive graph with tooltips
chart = alt.Chart(df).mark_point().encode(
    x=alt.X('Localidade:N', title='Região'),
    y=alt.Y('Pessoa com deficiência:Q', title='Total de Pessoas'),
    shape=formato_condicoes,
    size=alt.Size('Total:Q', title='Número de Pessoas'),
    tooltip=[
        alt.Tooltip('Localidade:N', title='Região'),
        alt.Tooltip('Total:Q', title='Total de Pessoas'),
        alt.Tooltip('Força de Trabalho:N', title='Força de Trabalho'),
        alt.Tooltip('Pessoa com deficiência:Q', title='Número de Pessoas')
    ]
).properties(
    title='Relação entre Número de Pessoas com Deficiência na Força de Trabalho e Total por Região Brasileira'
).configure_scale(
    bandPaddingInner=0.1
).configure_view(
    width=800  # Desired width of the graph in pixels
)

# Display the interactive graph
chart.interactive()
```

The result shows that there are always more people with disabilities outside the workforce than within it. In the North and Midwest, the difference is smaller.

In the Southeast, it's where there are the most people with disabilities outside the workforce, totaling 4,857. It's where there is the greatest disparity concerning people with disabilities in the workforce.

Variable "Locality":

* Graphic Encoding: X-Axis (horizontal position)
* Nature of Information: Nominal Categorical
(Geographical regions of Brazil, such as "North", "Northeast", etc.)

Variable "Workforce":

* Graphic Encoding: Point Shape (triangle for "In the workforce" and square for "Out of the workforce")
* Nature of Information: Nominal Categorical (Condition regarding the workforce, such as "In the workforce" or "Out of the workforce")

Variable "Person with disability":

* Graphic Encoding: Y-Axis (vertical position)
* Nature of Information: Discrete Numeric (Total number of people with disabilities)

Variable "Total":

* Graphic Encoding: Point Size
* Nature of Information: Discrete Numeric (Total number of people in each region)

## DATA VIS II
### Realizing the number of people with disabilities outside the workforce, I decided to understand in more depth the difference in presence or absence by education level.
### It was necessary to generate an interactive visualization with a dropdown so that the data would become clearer.

```python
import ipywidgets as widgets
from IPython.display import display

# Enable Altair rendering mode for Google Colab
alt.renderers.enable('colab')

# Function to update the graph according to the selection in the dropdown
def update_plot(nivel_instrucao):
    # Filter relevant columns for the selected education level
    colunas_selecionadas = [f'Pessoa {nivel_instrucao} com deficiência']
    dados_plot = tabela3[['Localidade', 'Condição em relação à força de trabalho'] + colunas_selecionadas]

    # Filter relevant rows (regions and conditions)
    filtro_linhas = dados_plot['Localidade'].isin(['Norte', 'Nordeste', 'Centro-Oeste', 'Sul', 'Sudeste']) & \
                    dados_plot['Condição em relação à força de trabalho'].isin(['Na força de trabalho', 'Fora da força de trabalho'])
    dados_plot = dados_plot[filtro_linhas]

    # Create a column for the total number of people per region
    dados_plot['Total'] = dados_plot.groupby('Localidade')[colunas_selecionadas[0]].transform('sum')

    # Transform the data into the appropriate format for Altair
    data = {
        'Localidade': dados_plot['Localidade'],
        'Força de Trabalho': dados_plot['Condição em relação à força de trabalho'],
        'Pessoa com deficiência': dados_plot[colunas_selecionadas[0]],
        'Total': dados_plot['Total']
    }
    df = pd.DataFrame(data)

    # Create an interactive graph with tooltips
    chart = alt.Chart(df).mark_bar().encode(
        x=alt.X('Localidade:N', title='Região'),
        y=alt.Y('Pessoa com deficiência:Q', title='Número de Pessoas'),
        color=alt.Color('Força de Trabalho:N', title='Força de Trabalho', scale=alt.Scale(domain=['Na força de trabalho', 'Fora da força de trabalho'], range=['orange', 'blue'])),
        tooltip=[
            alt.Tooltip('Localidade:N', title='Região'),
            alt.Tooltip('Total:Q', title='Total de Pessoas'),
            alt.Tooltip('Força de Trabalho:N', title='Força de Trabalho'),
            alt.Tooltip('Pessoa com deficiência:Q', title='Número de Pessoas')
        ]
    ).properties(
        title=f'Condição em Relação à Força de Trabalho de Pessoas Com Deficiência por {nivel_instrucao} e Região Brasileira'
    ).configure_axisX(
        labelAngle=-45
    ).configure_scale(
        bandPaddingInner=0.1
    ).configure_view(
        width=800  # Desired width of the graph in pixels
    )

    # Display the interactive graph
    display(chart)

# List of education levels
niveis_instrucao = ['sem instrução', 'nível básico', 'nível médio', 'nível superior']

# Create an interactive dropdown
dropdown = widgets.Dropdown(options=niveis_instrucao, description='Nível de Instrução:', style={'description_width': 'initial'})
output = widgets.interactive_output(update_plot, {'nivel_instrucao': dropdown})

# Display the interactive dropdown and graph
widgets.VBox([dropdown, output])
```

Remember that a total of 4,857 were out of the workforce only in the Southeast. Now, it's possible to know that, among these, 3,172 are people without education. And only 277 are people with disabilities and complete higher education.

The graph shows that only people with disabilities who have completed higher education have a greater chance of being employed than unemployed. This holds true for any region of Brazil.

Regarding people with disabilities and high school education, only in the North, South, and Midwest regions are there more people employed than unemployed. But the difference is very subtle.

* Nature of Information: Nominal Categorical (Geographical regions of Brazil, such as "North", "Northeast", etc.)

Variable "Workforce":

* Graphic Encoding: Bar Color (orange for "In the workforce" and blue for "Out of the workforce")
* Nature of Information: Nominal Categorical (Condition regarding the workforce, such as "In the workforce" or "Out of the workforce")

Variable "Person with disability":

## DATAVIS III
### Finally, the average salary analysis. Comparing also by educational level (visualizations available through the dropdown), this time the comparison was between people with and without disabilities. So, who earns more?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to update the graph according to the selected education level
def update_plot(change):
    nivel_instrucao = change['new']

    # Clear previous output
    clear_output(wait=True)

    # Filter relevant columns for the selected education level
    colunas_selecionadas = [f'Média - {nivel_instrucao}']
    dados_plot = tabela10[['Localidade', 'Existência de deficiência'] + colunas_selecionadas]

    # Filter relevant rows (regions and conditions)
    filtro_linhas = dados_plot['Localidade'].isin(['Norte', 'Nordeste', 'Centro-Oeste', 'Sul', 'Sudeste']) & \
                    dados_plot['Existência de deficiência'].isin(['Pessoa com deficiência', 'Pessoa sem deficiência'])
    dados_plot = dados_plot[filtro_linhas]

    # Transform the data into the appropriate format for plotting with Seaborn
    df = dados_plot.pivot_table(index=['Localidade', 'Existência de deficiência'], values=colunas_selecionadas[0]).reset_index()

    # Create the grouped bar plot with Seaborn
    sns.set_palette("pastel")
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(data=df, x='Localidade', y=colunas_selecionadas[0], hue='Existência de deficiência')

    # Configure the plot
    ax.set_xlabel('Região')
    ax.set_ylabel('Renda média mensal (em R$)')
    ax.set_title(f'Renda Média Mensal de Pessoas Com e Sem Deficiência por {nivel_instrucao} e Região Brasileira')
    ax.legend(title='Existência de deficiência')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=-45, ha='left')
    plt.tight_layout()

    # Display the plot
    display(dropdown_instrucao)
    plt.show()

# List of education levels
niveis_instrucao = ['Sem instrução e fundamental incompleto', 'Fundamental completo e médio incompleto', 'Médio completo e superior incompleto', 'Superior completo']

# Create radio buttons for the education levels
dropdown_instrucao = widgets.Dropdown(options=niveis_instrucao, description='Nível de Instrução:', style={'description_width': 'initial', 'width': '400px'}, layout={'width': '500px'})
dropdown_instrucao.observe(update_plot, names='value')

# Display the initial dropdown
update_plot({'new': niveis_instrucao[0]})
```

In the North, people with complete elementary education and incomplete high school, with or without disabilities, are the ones with more comparable salaries. In this situation, in the Northeast, the difference is a bit greater, but the average salary is also lower.

By the way, in the Northeast, the salary is always lower. Meanwhile, the South and Midwest have deeper pockets, especially for people with higher education, where the Southeast also stands out.
